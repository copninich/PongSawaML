### Scraping with BeautifulSoup Library

```python
def get_soup(url):
    with requests.get(url) as r:
        soup = BeautifulSoup(r.text, features='html.parser')
    return soup

def find_domain(url):
    counter = 0
    domain_name = ""
    for text in url:
        if counter <= 2:
            if text != "/":
                domain_name+= text
            else:
                counter += 1
                domain_name+= text
        else:
            return domain_name

def text_cleaning(context):
  context = re.sub(r'<.*?>','', context)
  context = ' '.join(context.split())
  context = re.sub(r',','',context)
  context = re.sub(r'----------------------------','',context)
  context = re.sub(' +', ' ', context)
  context = re.sub(r'\u200b','',context)
  return context
```

```python
def extract_kings(param):
  for king_name in param.find_all('h3'):
    if king_name not in list_of_king:
      list_of_king.append(king_name.text.strip())

def extract_context(param2):
  for context_timeline in param2.find_all('p'):
    list_of_annals_context.append(context_timeline.text.strip())
```

```python
MAX_SCRAPING = 1
urls = navigation_df_new.url.tolist()

def get_post_head_desc_append(url, res, res2):
    soup = get_soup(url)
    res.append({'king_name': extract_kings(soup)})
    res2.append({'annals': extract_context(soup)})

def scrape_multithread(urls, scrape_func):
    threads = min(MAX_SCRAPING, len(urls))
    with concurrent.futures.ThreadPoolExecutor(max_workers=threads) as executor:
        executor.map(scrape_func, urls)
```

```python
res = []
res2 = []
list_of_king=[]
list_of_annals_context = []
scrape_multithread(urls, partial(get_post_head_desc_append, res=res ,res2=res2))
```
